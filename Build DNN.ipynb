```python
def split_data(df,test_size=0.4,val_test=0.5):
    X, Y = df[:, :-1], df[:, -1]
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size,random_state=1)
    X_vali, X_test, Y_vali, Y_test = train_test_split(X_test, Y_test, test_size=val_test,random_state=1)
    data = [X_train,Y_train,X_vali,Y_vali,X_test,Y_test]
    data ={"X_train":X_train,
           "Y_train":Y_train,
           "X_vali":X_vali,
           "Y_vali":Y_vali,
           "X_test":X_test,
           "Y_test":Y_test
    }
    return data
```

```python
data = split_data(normalized)
data["X_train"],data["Y_test"]
```

```python
def BGM(X_train,Y_train,X_vali,Y_vali,X_test,epochs):
    model = Sequential()
    model.add(Dense(350, input_dim=5,activation = "relu")) #Layer 1
    model.add(Dense(350,activation = "relu")) #Layer 2
    model.add(Dense(350,activation = "relu")) #Layer 2
    model.add(Dense(350,activation = "relu")) #Layer 2
    model.add(Dense(350,activation = "relu")) #Layer 3
    model.add(Dense(1,activation = "linear")) #Output
    model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mse'])
    model.fit(X_train, Y_train, epochs=epochs,validation_data=(X_vali,Y_vali))
    result = model
    return result
```

```python
predeiction = BGM(data["X_train"],data["Y_train"],data["X_vali"],data["Y_vali"],data["X_test"],1000)
```

```python
X_test = predeiction.predict(data["X_test"])
X_test
```
